```python
trp_groups = stop_times_tbl.groupby('trip_id')
last_first_stp = trp_groups['stop_sequence'].agg(['max', 'min'])
last_first_stp.head()
```

```python
def runtime(df, trip_id, max_stop, min_stop):
    trip = df[df['trip_id'] == trip_id]
    last_stop_t = trip[trip['stop_sequence'] == max_stop]['arrival_time'].iloc[0]
    first_stop_t = trip[trip['stop_sequence'] == min_stop]['arrival_time'].iloc[0]
    
    times = [list(map(int, first_stop_t.split(':'))), list(map(int, last_stop_t.split(':')))]
    time_deltas = [timedelta(hours=time[0], minutes=time[1], seconds=time[2]) for time in times]
    time_diff = time_deltas[1] - time_deltas[0]
    
    return time_diff.seconds
last_first_stp.head()
```
